```python
import pandas as pd
import pickle
```

```python
data:dict = pickle.load(open("dataaaaa.pkl","rb"))
```

```python
for d in data:
    print(d)
```

```python
for model in data:
    print(model)
```

```python
to_excel={}
for fs in data:
    
    for score in data[fs]:
        model = score["model"]
                
        for metric in score:
            if metric=="model":
                continue
            if not "model" in to_excel:
                to_excel["model"]=[]
            if not "selector" in to_excel:
                to_excel["selector"]=[]
            
            to_excel["selector"].append(fs)
            to_excel["model"].append(model)
            if "score" not in to_excel:
                to_excel["score"]=[]
            to_excel["score"].append(metric)
            for result in score[metric]:

                if result not in to_excel:
                    to_excel[result]=[]
                to_excel[result].append(score[metric][result])
```

```python
to_excel
```

```python
df = pd.DataFrame(to_excel)
```

```python
df.to_excel("scores.xlsx")
```
